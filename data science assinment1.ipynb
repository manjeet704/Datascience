Q3. Explain the use of the following operators using an example:

(i)	/

(ii)	% 

(iii)	//

(iv)	**

Ans3:(i) / Operator:
The "/" operator is the division operator in most programming languages. It is used to divide one numerical value by another. The result of the division operation is the quotient or the ratio of the two numbers.

Example:
a = 10
b = 3
result = a / b
print(result)

the variables a and b hold the values 10 and 3 respectively. The line result = a / b performs the division operation, dividing a by b. The result will be a floating-point number, which in this case would be 3.3333. Finally, the print(result) statement will output the result as 3.3333.

(ii) % Operator:
The "%" operator is called the modulo operator or the remainder operator. It calculates the remainder of the division operation between two numbers. It is often used to check if a number is divisible by another number.
a = 10
b = 3
result = a % b
print(result)
result = a % b calculates the remainder when a is divided by b. The result will be the remainder of 10 divided by 3, which is 1.

The "//" operator is the floor division operator. It performs division between two numbers and rounds the result down to the nearest integer, dropping any decimal part. It is useful when you want to obtain the integer quotient of a division operation.
a = 10
b = 3
result = a // b
print(result)
result = a // b performs the floor division of a by b. The result will be the integer quotient of the division, which in this case is 3

(iv)
The "**" operator represents the exponentiation operator. It is used to raise a number to the power of another number. The operator takes the base and the exponent as operands.
base = 2
exponent = 3
result = base ** exponent
print(result)
result = base ** exponent calculates 2 raised to the power of 3. The result will be 8. The output of print(result) will be 8.

Q7. What do you understand about mutable and immutable data types? Give examples for both showing
this property.
Ans:
Immutable data types:
Immutable data types are those whose values cannot be changed after they are created. Any operation that appears to modify an immutable object actually creates a new object with the modified value. Once an immutable object is created, its value remains the same throughout its lifetime.

Example of immutable data types:
Integer: In Python, integers are immutable. Once an integer object is created, its value cannot be changed. Any operation that modifies an integer, such as addition or subtraction, creates a new integer object.
x = 5
print(x)  # Output: 5

x = x + 1
print(x)  # Output: 6

String: Strings in many programming languages, including Python, are immutable. Once a string is created, its contents cannot be modified. Any operation that appears to modify a string actually creates a new string object.

s = "Hello"
print(s)  # Output: Hello

s = s + " World"
print(s)  # Output: Hello World

Mutable data types:
Mutable data types, on the other hand, allow modifications to their values after they are created. This means the object can be modified in-place without creating a new object with each modification.

Example of mutable data types:
List: Lists in Python are mutable. You can modify the elements of a list without creating a new list.
my_list = [1, 2, 3]
print(my_list)  # Output: [1, 2, 3]

my_list[1] = 4
print(my_list)  # Output: [1, 4, 3]


Dictionary: Dictionaries in Python are mutable. You can modify the key-value pairs of a dictionary without creating a new dictionary.
my_dict = {"name": "John", "age": 25}
print(my_dict)  # Output: {'name': 'John', 'age': 25}

my_dict["age"] = 26
print(my_dict)  # Output: {'name': 'John', 'age': 26}
In this example, we modify the value associated with the "age" key directly. The original dictionary is modified in-place, and there is no need to create a new dictionary.

In summary, immutable data types do not allow modification of their values, whereas mutable data types allow modifications in-place.
